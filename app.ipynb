Importing the Dependencies

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Import boston house price dataset from sklearn.dataset

```python
house_data=sklearn.datasets.load_boston()
```

```python
print(house_data)
```

```python
house_dataframe=pd.DataFrame(house_data.data,columns=house_data.feature_names)
```

```python
# print first 5 rows of data
house_dataframe.head()
```

```python
house_dataframe['price']=house_data.target
# note => price is in division of 1000 dollars
```

```python
# print first 5 rows of data after adding price
house_dataframe.head()
```

```python
# check the shape of daatframe
house_dataframe.shape
```

```python
# check general info about data
house_dataframe.describe()
```

```python
# check for missing values in datafram
house_dataframe.isnull().sum()
```

None of data is null we dont have remove to insert value there before givong this data to our ML model

Now lets understand the relationship between the columns
1) Positive Correlation
2) Negative Correlation

```python
correlation=house_dataframe.corr()
# corr() method =>
#  Compute pairwise correlation of columns, excluding NA/null values.
correlation
```

By seeing this it is diffcult to understand whats going on, So lets Plot this data using matplot

```python
# lets plot a head map to better understand the data
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True, square=True, annot=True, fmt='.1f', annot_kws={'size':8})
# sqaure to make tiles square
# annot to print the digits in titles
# fmt to make decimal to limit till first digit ex. 0.1
```

Split Data in training and test

```python
X=house_dataframe.drop(['price'],axis=1)
Y=house_dataframe['price']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
# loading the model
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

Evaluation the model

```python
# prediction using training data
training_data_prediction=model.predict(X_train)
print(training_data_prediction)
```

```python
""" Accuracy with training data """

# R Square Error =>
#  R-squared represents the fraction of variance of the actual value of the response variable captured by the regression model

train_r_score=metrics.r2_score(Y_train,training_data_prediction)
print('R Square Error', train_r_score)

# Mean Absolute Error =>
#  Absolute Error is the amount of error in your measurements. It is the difference between the measured value and true value.

train_mean_abs_score=metrics.mean_absolute_error(Y_train,training_data_prediction)
print('Mean Absolute Error', train_mean_abs_score)
```

Visualizing the actual Prices and predicted prices (Training)

```python
plt.scatter(Y_train,training_data_prediction, marker='^')
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price in Training")
plt.show()
```

```python
# prediction using testing data
testing_data_prediction=model.predict(X_test)
print(testing_data_prediction)
```

```python
""" Accuracy with testing data """

# R Square Error =>
#  R-squared represents the fraction of variance of the actual value of the response variable captured by the regression model

test_r_score=metrics.r2_score(Y_test,testing_data_prediction)
print('R Square Error', test_r_score)

# Mean Absolute Error =>
#  Absolute Error is the amount of error in your measurements. It is the difference between the measured value and true value.

test_mean_abs_score=metrics.mean_absolute_error(Y_test,testing_data_prediction)
print('Mean Absolute Error', test_mean_abs_score)
```

Visualizing the actual Prices and predicted prices (Testing)

```python
plt.scatter(Y_test,testing_data_prediction, marker='o')
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price in Testing")
plt.show()
```
